```python
def get_teamcode(teamstr, teamcode="", allteam=False, allcode=False):
    teamdict = {"巨人":"g",\
                "阪神":"t",\
                "中日":"d",\
                "DeNA":"yb",\
                "横浜":"yb",\
                "広島":"c",\
                "ヤクルト":"s",\
                "ソフトバンク":"h",\
                "ロッテ":"m",\
                "西武":"l",\
                "楽天":"e",\
                "日本ハム":"f",\
                "オリックス":"bs"}
    if allcode:
        return list(teamdict.values())
    if allteam:
        return list(teamdict.keys())
    if teamcode:
        return [k for k, v in teamdict.items() if v == teamcode][0]
    return teamdict[teamstr]
```

```python
#urlをリスト形式で取得
df_all = []
#各要素に各年のデータが入る(執筆時点では2020年が最新)
# 2020年から2008年まで
years = range(20,8,-1)
urls = []
#URLを入力：最新年度だけ命名規則が違う
for year in years:
    urls.append('http://baseball-data.com/'+ "{0:02d}".format(year)+'/stats/hitter-all/avg-1.html')
year=20
#--------------------------データをURLから取得-------------------------------------------
for url in urls:
    print('取得URL：'+url)
    df = pd.io.html.read_html(url)
    df = df[0]
    df.columns = [c[0] for c in df.columns]
    df_all.append(df)
    file_path="npb_hitter_"+str(year)+".csv"
    df.to_csv(file_path,encoding="utf-8",index=None)
    year=year - 1 
```

```python
df_all = []
#各要素に各年のデータが入る(執筆時点では2020年が最新)
# 2020年から2008年まで
years = range(20,8,-1)
teamcodes = get_teamcode("", allcode=True)
#URLを入力：最新年度だけ命名規則が違う
for year in years:
    df_allteam = pd.DataFrame()
    for teamcode in teamcodes:
        url = 'http://baseball-data.com/'+ "{0:02d}".format(year)+'/ranking-salary/'+teamcode+'/h.html'
        print('取得URL：'+url)
        df = pd.io.html.read_html(url)
        df = df[0]
        df.columns = [c[0] for c in df.columns]
        df = df.drop("順位", axis=1)
        df_allteam = pd.concat([df_allteam, df])

    df_all.append(df_allteam)
    file_path="npb_hitter_"+str(year)+".csv"
    df_orig=pd.read_csv(file_path,encoding="utf-8")
    df_merged = pd.merge(df_orig, df_allteam, how="outer", on="選手名")

    file_path_new="npb_hitter_"+str(year)+"_allinfo.csv"
    df_merged.to_csv(file_path_new,encoding="utf-8",index=None)
```

```python
df_all = []
for year in years:
    file_path="npb_hitter_"+str(year)+"_allinfo.csv"
    df=pd.read_csv(file_path,encoding="utf-8")
    df_all.append(df)
    print('読み込んだファイル：'+file_path)
    df = df[~df.duplicated()]

    # 単位を取り除く
    df = df.replace("(.*)万円", {'年俸(推定)':r'\1'}, regex=True)
    df = df.replace("-", {'年俸(推定)':None}, regex=True)
    df = df.replace("(.*)歳", {'年齢':r'\1'}, regex=True)
    df = df.replace("(.*)年", {'年数':r'\1'}, regex=True)
    df = df.rename(columns={'年俸(推定)':'年俸(万円)'})
    df = df.replace("(\d*),(\d{3})", {'年俸(万円)':r'\1\2'}, regex=True)
    df = df.replace("(.*)[都府県]", {'出身地':r'\1'}, regex=True)
    df = df.replace("京", {"出身地":"京都"}, regex=True)
    df["年俸(万円)"] = df["年俸(万円)"].astype('float').astype('Int64')
    df["年数"] = df["年数"].astype('float').astype('Int64')
    df["年齢"] = df["年齢"].astype('float').astype('Int64')

    # 外国出身
    df["外国出身"] = ~df["出身地"].isin(prefecs)

    df.to_csv(file_path,encoding="utf-8",index=None)
```

```python
prefecs = ["北海道","青森","岩手","宮城","秋田","山形","福島","茨城","栃木","群馬","埼玉","千葉","東京","神奈川","新潟","富山","石川","福井","山梨","長野","岐阜","静岡","愛知","三重","滋賀","京都","大阪","兵庫","奈良","和歌山","鳥取","島根","岡山","広島","山口","徳島","香川","愛媛","高知","福岡","佐賀","長崎","熊本","大分","宮崎","鹿児島","沖縄"]
```
